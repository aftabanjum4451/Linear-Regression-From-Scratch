```python
# import data
import pandas as pd
df=pd.read_csv('/content/gdrive/MyDrive/pytorch learning and project/liner regression from scratch/airfoil_self_noise.csv',header=None)
```

```python
df.columns=['Freq','Angle','Chord_length','Free_stream','d_thickness','targets']
df.shape
```

```python
df.head()
```

```python
#check null values
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
import torch 
import numpy as np
```

```python
# Input array
inputs = np.array(df.drop('targets', 1), dtype='float32')
targets = np.array(df['targets'], dtype='float32')
```

```python
inputs.shape,targets.shape
```

```python
# Convert inputs and targets to tensors
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
import torch.nn as nn
from torch.utils.data import TensorDataset
```

```python
# Define dataset
train_ds = TensorDataset(inputs, targets)
train_ds[0:3]
```

## Data loader module

```python
from torch.utils.data import DataLoader
# Define data loader
batch_size = 40
train_dl = DataLoader(train_ds, batch_size, shuffle=True)
```

lets see the train_dl created by data loader

```python
for xb, yb in train_dl:
    print(xb)# trainning the batches 
    print(yb)# targets of batches
    break
```

## nn.Linear

```python
# Define model, and initialize the weights 
model = nn.Linear(5, 1)
print(model.weight)
print(model.bias)
```

```python
# Parameters: check all the model parameters
list(model.parameters())
```

```python
# Generate predictions
preds = model(inputs)
preds
```

## loss function

```python
# Import nn.functional
import torch.nn.functional as F
```

```python
# Define loss function
loss_fn = F.mse_loss
pred=model(inputs)
loss = torch.sqrt(loss_fn(pred.squeeze(), targets))
print(loss)
```

# Optimizer 

```python
# Define optimizer
opt = torch.optim.SGD(model.parameters(), lr=1e-6)
```

# Model Training

```python
# Utility function to train the model
def fit(num_epochs, model, loss_fn, opt, train_dl):
    
    # Repeat for given number of epochs
    for epoch in range(num_epochs):
        
        # Train with batches of data
        for xb,yb in train_dl:
            
            # 1. Generate predictions
            pred = model(xb)
            # 2. Calculate loss
            loss = torch.sqrt(loss_fn(pred.squeeze(), yb))
            #print(loss)
            
            # 3. Compute gradients
            loss.backward()
            
            # 4. Update parameters using gradients
            opt.step()
            
            # 5. Reset the gradients to zero
            opt.zero_grad()
        
        # Print the progress
        if (epoch+1) % 10 == 0:
            print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```

```python
fit(1000, model, loss_fn, opt, train_dl)
```
